```python
import spacy

# Load model
nlp = spacy.load("en_core_web_sm")

# Example text
text = "Tesla Inc is going to acquire Twitter Inc for $45 billion"

# Process text
doc = nlp(text)

print("\n--- Named Entities ---\n")

for ent in doc.ents:
    print(f"{ent.text} | {ent.label_} | {spacy.explain(ent.label_)}")
```

```python
from spacy import displacy

displacy.render(doc, style="ent", jupyter=True)
```

```python
text = input("Enter your sentence: ")

doc = nlp(text)

print("\nEntities Found:\n")

for ent in doc.ents:
    print(f"{ent.text} --> {ent.label_} ({spacy.explain(ent.label_)})")
```

```python
from spacy.tokens import Span

doc = nlp("Tesla is going to acquire Twitter for $45 billion")

# Create custom entities
ent1 = Span(doc, 0, 1, label="ORG")
ent2 = Span(doc, 5, 6, label="ORG")

doc.set_ents([ent1, ent2], default="unmodified")

print("\nAfter Custom Entities:\n")

for ent in doc.ents:
    print(ent.text, "|", ent.label_)
```
